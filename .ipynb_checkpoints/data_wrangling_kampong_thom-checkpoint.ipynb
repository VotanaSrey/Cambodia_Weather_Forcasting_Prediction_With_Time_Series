```python
# importing the required modules
import glob
# Import the neccessary modules for data manipulation and visual representation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplot
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
path='../Cambodia_Weather_Forcasting_Prediction_With_Time_Series/datasets/raw_layer/kampong_thom'
file_list = glob.glob(path + "/*.xlsx")

excl_list = []

for file in file_list:
    excl_list.append(pd.read_excel(file))
excl_merged = pd.DataFrame()

for excl_file in excl_list:
    excl_merged = excl_merged.append(excl_file, ignore_index=True)

excl_merged
```

```python
pd.set_option('display.max_columns', None)

data = excl_merged
data.head()
```

```python
#since there are some Unnecessary colums appear from merged sooo clean them out
columns = ['CE SAIN Weather Station', 'Export settings name']
data.drop(columns, inplace=True, axis=1)
data.head()
```
